# Line Graph

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
gas = pd.read_csv("gas_prices.csv")
```

```python
gas.head()
```

```python
print(gas.Year)
```

```python
#print(gas.Year[0:5])
print(gas.Year[::3])
```

```python
plt.figure(figsize=(8,5))
plt.plot(gas.Year,gas.USA,'o-',color='red',label='USA',ms=5)
plt.plot(gas['Year'],gas.Canada,'bo-',label='Canada',ms=5)
plt.xlabel("USA vs Canada gas prices Year")
plt.ylabel("Dollars/Gallon")
plt.title("Gas prices overtime in USD",fontdict={'fontweight':'bold','fontsize':15})
#plt.xticks(gas.Year)
plt.xticks(gas.Year[::3].tolist()+[2011])

plt.legend()
plt.savefig("Gas_Price_Figure.png",dpi=300)
plt.show()
```

```python
plt.figure(figsize=(8,5))

countries_to_look_at = ['Australia','USA','Canada','South Korea']
for country in gas:
    #if country in countries_to_look_at:
    if country!='Year':
        #print(country)
        plt.plot(gas.Year,gas[country],'o-',label=country,ms=2)
        
plt.xlabel("USA vs Canada gas prices Year")
plt.ylabel("Dollars/Gallon")
plt.title("Gas prices overtime in USD")
#plt.xticks(gas.Year)
plt.xticks(gas.Year[::3])

plt.legend(ncol=3)
plt.show()
```

# Pie Chart

```python
for col in fifa.columns:
    print(col,end= ' ')
print("\n\n")
col_list = fifa.columns.tolist()
print(col_list,end=" ")
```

```python
fifa['Preferred Foot'].value_counts()
```

```python
left = fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
left
```

```python
right = fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]
right
```

```python
labels = ['Left','Right']
colors = ['y','#f5429e']
plt.pie([left,right],labels=labels,colors=colors,autopct='%.2f %%') #provide values as parameters
plt.title("Foot preference of FIFA players")
plt.show()
```

```python
#Using Weight column
fifa.Weight = [x.strip('lbs') if type(x)==str else x for x in fifa.Weight]
#fifa.Weight
fifa.Weight[0] # value is still in string
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
fifa.Weight[0]
```

```python
max_value = fifa['Weight'].max()
print(max_value)
min_value = fifa['Weight'].min()
print(min_value)
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

plt.style.use('ggplot')

light = fifa.loc[fifa.Weight < 135].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 135) & (fifa.Weight < 150)].count()[0] 
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 190)].count()[0]
heavy = fifa.loc[fifa.Weight >= 190].count()[0]
#heavy
weights = [light,light_medium,medium,medium_heavy,heavy]
labels = ['Under 135','135-150','150-175','175-190','Over 190']
#explode =(.1,.1,0,.1,.1)
explode = (.4,.2,0,0,.4)

plt.title("Weight distribution of Fifa players (in lbs)")
plt.pie(weights,labels=labels,autopct='%.2f %%',pctdistance=0.75,explode=explode)#pct=percentage distance

plt.show()
```

```python
plt.figure(figsize=(5,6))
plt.style.use('default')
barcelona = fifa.loc[fifa.Club=='FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club=='Real Madrid']['Overall']
revs = fifa.loc[fifa.Club=='New England Revolution']['Overall']

labels = ['FC Barcelona','Real Madrid','NE Revolution']
boxes = plt.boxplot([barcelona,madrid,revs],labels=labels,patch_artist=True,medianprops={'linewidth':2})#Patch_artist: set true, when gets error while using facecolor property 

for box in boxes['boxes']:
    print(box)
    #Set edge color
    box.set(color='#4286f4',linewidth=2)
    #Change fill color
    box.set(facecolor='#e0e0e0')
        
plt.title("Professional Soccer Team Comparison")
plt.ylabel("Fifa Overall rating")
plt.show()
```
