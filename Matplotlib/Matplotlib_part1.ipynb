# A More Advanced API

For more complicated plots, such as
- Multiple subplots on 1 figure
- Conveniently adding text to plots
- Creating animations

one needs to use a slightly more advanced API

```python
#Single Axes in one figure
fig, ax = plt.subplots(1,1,figsize=(12,4))
ax.plot(x,y,'o--')
ax.set_xlabel('Time[s]')
ax.set_ylabel('Voltage[v]')
```

```python
#Multiple Axes in one figure
fig, axes = plt.subplots(3,2,figsize=(12,12))
ax = axes[0][0]
ax.plot(x,y,'o--')
ax.set_xlabel('Time[s]')
ax.set_ylabel('Voltage[v]')
ax.set_title('Line Plot')
#ax.text(0.1,0.1,'text here')
ax.text(0.05,0.05,'text here',transform=ax.transAxes)
ax.tick_params(axis='both',labelsize=10) # to change the fontsize of tick numbers

ax = axes[1][1]
ax.hist(res,bins=30,density=True,histtype='step',label='comp1')
ax.set_xlabel('Time[s]')
ax.set_ylabel('Voltage[v]')
ax.set_title('Histogram')
ax.legend()
plt.show()
```

Create the following plot

```python
res_a1 = 0.2*np.random.randn(1000)+0.4
res_b1 = 0.25*np.random.randn(1000)+0.4
res_a2 = 0.21*np.random.randn(1000)+0.3
res_b2 = 0.22*np.random.randn(1000)+0.3
```

```python
#Quick histogram plotting
plt.hist(res_a1,bins=100)
plt.hist(res_b1,bins=100)
plt.grid(linestyle='--')
plt.show()
```

good habit to come up with complicated strings beforehand

```python
textstr1 = '\n'.join((r'$\sigma_a=%.4f$' % (np.std(res_a1)),r'$\sigma_b=%.4f$' % (np.std(res_a1))))
textstr2 = '\n'.join((r'$\sigma_a=%.4f$' % (np.std(res_a2)),r'$\sigma_b=%.4f$' % (np.std(res_b2))))
```

```python
textstr1
```

```python
textstr2
```

Make plot

```python
fig, axes = plt.subplots(1,2,figsize=(10,3.5))
ax = axes[0]
ax.hist(res_a1,bins=30,density=True,histtype='step',label='a')
ax.hist(res_b1,bins=30,density=True,histtype='step',label='b')
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
ax.legend(loc='upper right',frameon=True,edgecolor='black')
ax.text(0.05,0.81,textstr1,transform=ax.transAxes,bbox=dict(facecolor='white',edgecolor='black'))

ax = axes[1]
ax.hist(res_a2,bins=30,density=True,histtype='step',label='a')
ax.hist(res_b2,bins=30,density=True,histtype='step',label='b')
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
ax.legend(loc='upper right')
ax.text(0.05,0.81,textstr2,transform=ax.transAxes)

plt.savefig('plot.png',dpi=200)
plt.show()
```

```python
fig, axes = plt.subplots(1,2,figsize=(10,3.5))
ax = axes[0]
ax.hist(res_a1,bins=30,density=True,histtype='step',label='a',color='blue')
ax.hist(res_b1,bins=30,density=True,histtype='step',label='b',color='red')
ax.set_ylabel("Density",fontsize=20)
ax.legend(loc='upper right',frameon=True,edgecolor='black')
ax.text(0.05,0.81,textstr1,transform=ax.transAxes,bbox=dict(facecolor='white',edgecolor='black'))
fig.text(0.5,-0.04,'$\Delta E$ [Joules]',ha='center',fontsize=20)#ha=horizontal alignment
ax.tick_params(axis='both',labelsize=20)
ax.grid(color = 'grey', linestyle = '--', linewidth = 0.5)
ax.set_title('Trial 1')

ax = axes[1]
ax.hist(res_a2,bins=30,density=True,histtype='step',label='a',color='blue')
ax.hist(res_b2,bins=30,density=True,histtype='step',label='b',color='red')
ax.legend(loc='upper right')
ax.text(0.05,0.81,textstr2,transform=ax.transAxes)#The coordinate system of the Axes; (0,0) is bottom left of the axes, and (1,1) is top right of the axes.
ax.tick_params(axis='both',labelsize=20)
ax.grid(color = 'grey', linestyle = '--', lw = 0.5)
ax.set_title('Trial 2')
#plt.yticks(np.arange(0, 2, 0.5))

plt.show()
```

# Stream Plots

For plotting vector plots

```python
_ = np.linspace(-3,3,100)
X,Y = np.meshgrid(_,_)
U = -1 - X**2 + Y
V = 1 + X - Y**2
speed = np.sqrt(U**2 + V**2)
```

```python
fig, axes = plt.subplots(2,2,figsize=(8,8))
ax = axes[0][0]
ax.streamplot(X,Y,U,V)

ax = axes[0][1]
ax.streamplot(X,Y,U,V,color=speed)

ax = axes[1][0]
lw = 5*speed/speed.max()
ax.streamplot(X,Y,U,V,linewidth = lw)

ax = axes[1][1]
seedpoints = np.array([[0,1],[1,0]])
ax.streamplot(X,Y,U,V,start_points=seedpoints)
```

# Animations

Animations require relating the time of your problem to the frame of the animation, then specifying the frames per second
- below is a sin wave moving at 3m/s. Translate this into an animation

```python
def f(x,t):
    return np.sin(x-3*t)
x = np.linspace(0,10*np.pi,1000)
```

```python
plt.plot(x,f(x,0))
plt.plot(x,f(x,0.2))
plt.plot(x,f(x,0.4))
plt.plot(x,f(x,0.6))
```

Create animation

```python
from matplotlib import animation
from matplotlib.animation import PillowWriter
```

```python
fig,ax = plt.subplots(1,1,figsize=(8,4))
ln1, = plt.plot([],[])
time_text = ax.text(0.65,0.95,'',fontsize=15,transform=ax.transAxes,bbox=dict(facecolor='white',edgecolor='black'))
ax.set_xlim(0,10*np.pi)
ax.set_ylim(-1.5,1.5)
def animate(i):
    ln1.set_data(x,f(x,1/50*i))
    time_text.set_text('t={:.2f}'.format(i/50))
ani = animation.FuncAnimation(fig,animate,frames=240,interval=50)
ani.save('ani.gif',writer='pillow',fps=50,dpi=100)
```

The animation function can be used in other ways too. One can create 3D rotating gifs to surfaces

```python
_ = np.linspace(-1,1,100)
x,y = np.meshgrid(_,_)
z = x**2+x*y
```

```python
fig,ax = plt.subplots(subplot_kw={'projection':'3d'})
ax.plot_surface(x,y,z,cmap='coolwarm',linewidth=0,antialiased=False)
ax.view_init(elev=10,azim=0)

def animate(i):
    ax.view_init(elev=10,azim=3*i)
    
ani = animation.FuncAnimation(fig,animate,frames=240,interval=50)
ani.save('ani1.gif',writer='pillow',fps=50,dpi=100)
```
