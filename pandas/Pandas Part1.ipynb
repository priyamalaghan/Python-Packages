# To subset rows using loc and iloc

loc gets rows (and/or columns) with particular labels.

iloc gets rows (and/or columns) at integer locations.

```python
gapminder.loc[0]
```

```python
gapminder.iloc[0]
```

```python
#gapminder.loc[-1] #key error:-1 # -1 label is not there #loc do string match
gapminder.iloc[-1]
```

```python
gapminder.loc[[0,1,8,15,2]]
```

```python
gapminder.iloc[[0,1,8,15,2,-1]]
```

```python
gapminder.loc[3:5]
```

```python
gapminder.iloc[3:5]
```

```python
subset = gapminder.loc[:, ['year','country']]
subset
```

```python
subset = gapminder.loc[:, :]
subset
```

```python
subset = gapminder.loc[3:6, ['year','country']]
subset
```

```python
#subset = gapminder.iloc[3:6, ['year','country']] #.iloc requires numeric indexers, got ['year' 'country']
subset = gapminder.iloc[3:6, [2,4]] 
subset
```

# Groupby:It allows you to split your data into separate groups to perform computations for better analysis

```python
graph = gapminder.groupby('year')['life_exp'].mean()
graph
```

```python
import matplotlib.pyplot as plt
graph = gapminder.groupby('year')['life_exp'].mean()
plt.plot(graph)
```

```python
gapminder.groupby(['country','year'])['life_exp'].mean()
```

```python
gapminder.groupby(['year','country'])['life_exp'].mean()
```

```python
import numpy as np
gapminder.groupby('year')['life_exp'].agg(np.std)
```

```python
gapminder.groupby(['year','continent'])[['life_exp','gdp_cap']].agg(np.mean)
```

```python
gapminder.groupby(['year','continent'])[['life_exp','gdp_cap']].agg(np.mean).reset_index() #toget dataframe back
```
