```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.columns
```

```python
#plt.plot(tips.tip)
tips.tip.plot()#default kind = line
```

```python
tips.tip.plot(kind='hist')
plt.show()
```

```python
cts = tips.smoker.value_counts()
cts
```

```python
cts.plot(kind='bar')
```

```python
sns.countplot(x='smoker', data=tips)
```

```python
sns.displot(tips.total_bill,kde=True)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',fit_reg=False)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',col='smoker')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',col='smoker',row='day')
```

```python
facet = sns.FacetGrid(tips,col='time',row='smoker',hue='sex')
facet.map(plt.scatter,'total_bill','tip')
```
