# Line Chart

```python
#A line chart displays information as a series of data points or markers, connected by straight lines
#one can customize the shape, size, color and aesthetic elements of the markers and lines for better visual clarity
```

```python
#yield of apples (tons per hectare) over 6 years
yield_apples = [0.895,0.91,0.919,0.926,0.929,0.931]
plt.plot(yield_apples,marker='.')
```

plotting multiple lines

```python
years = range(2000,2009)
apples = [0.895,0.91,0.919,0.926,0.929,0.931,0.934,0.936,0.937]
oranges = [0.926,0.941,0.930,0.923,0.918,0.908,0.907,0.904,0.901]
```

```python
plt.plot(years,apples)
plt.plot(years,oranges)
plt.xlabel("Year")
plt.ylabel("Yield (tons per hectare)")
```

Chart title and legend

```python
plt.plot(years,apples,label='Apple')
plt.plot(years,oranges,label='Orange')
plt.xlabel("Year")
plt.ylabel("Yield (tons per hectare)")
plt.title('Crop yields in kanto')
plt.legend()
```

```python
plt.plot(years,apples)
plt.plot(years,oranges)
plt.xlabel("Year")
plt.ylabel("Yield (tons per hectare)")
plt.title('Crop yields in kanto')
plt.legend(['Apples','Oranges'])
```

Line Markers and styling

```python
#color or c, linestyle or ls, linewidth or lw, markersize or ms, markeredgecolor or mec, markeredgewidth or mew, markerfacecolor or mfc
plt.plot(years,apples,label='Apple',marker='o',c='r',ls='--',lw=2,ms=6,mec='black',mew=1,mfc='y')
plt.plot(years,oranges,label='Orange',marker='s',c='b',ls='-.',lw=2,ms=6,mec='black',mew=1,mfc='y',alpha=0.5)
plt.xlabel("Year")
plt.ylabel("Yield (tons per hectare)")
plt.title('Crop yields in kanto')
plt.legend()
```

```python
# fmt : shorthand fro specifying the ls, marker and line color
# fmt = '[marker][line][color]'
plt.plot(years,apples,'s-b',label='Apple')
plt.plot(years,oranges,'o--r',label='Orange')
plt.xlabel("Year")
plt.ylabel("Yield (tons per hectare)")
plt.title('Crop yields in kanto')
plt.legend()
```

# Scatter Plot

```python
#The values of 2 variables are plotted as points on a 2 dimensional grid.
#Additionally, one can also use a third variable to determine the size or color of the points
flower_df = sns.load_dataset('iris')
flower_df.head()
```

```python
flower_df.species.unique()
```

```python
plt.plot(flower_df.sepal_length,flower_df.sepal_width)
```

```python
sns.scatterplot(x=flower_df.sepal_length,y=flower_df.sepal_width)
```

Adding Hues

```python
#Adding hues  makes the plot more informative
#Sesota species have a smaller sepal length but higher sepal width
#While the opposite holds true for the virginica species
plt.figure(figsize=(12,5))
plt.title("Sepal Dimensions")
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=flower_df,s=80)#s:size of hue marker
plt.legend(loc='upper right',ncol=3)
```

# Histogram

```python
#A histogram represents the distribution of data by forming bins along the range of the data and
#then drawing bars to show the number of observations that fall in each bin
```

```python
flower_df.head()
```

```python
flower_df.describe()
```

```python
plt.title("Distribution of sepal width")
plt.hist(flower_df.sepal_width)
```

controlling the size and number of bins

```python
plt.title("Distribution of sepal width")
plt.hist(flower_df.sepal_width,bins=5)
```

```python
import numpy as np
#Specifying the boundaries of each bin
plt.hist(flower_df.sepal_width,bins=np.arange(2,5,0.25))
```

```python
#Bins of unequal size
plt.hist(flower_df.sepal_width,bins=[1,3,4,4.5])
```

# Multiple Histograms

```python
setosa_df = flower_df[flower_df.species == 'setosa']
versicolor_df = flower_df[flower_df.species == 'versicolor']
virginica_df = flower_df[flower_df.species == 'virginica']
```

```python
plt.hist(setosa_df.sepal_width,alpha=0.4,bins=np.arange(2,5,0.25))
plt.hist(versicolor_df.sepal_width,alpha=0.2,bins=np.arange(2,5,0.25))
#plt.hist(virginica_df.sepal_width,alpha=0.2,bins=np.arange(2,5,0.25))

plt.legend(['Setosa','Versicolor','Virginica'])
```

```python
# We can also stack multiple histograms on top of one another
plt.title('Distribution of Sepal Width')
plt.hist([setosa_df.sepal_width,versicolor_df.sepal_width,virginica_df.sepal_width],bins=np.arange(2,5,0.25),stacked=True,alpha=0.5)
plt.legend(['Setosa','Versicolor','Virginica'])
```

# Bar Chart

```python
#Bar charts are quite similar to line charts i.e they show a sequence of values
#however a bar is shown for each value, rather than points connected by lines
years = range(2000,2006)
apples = [0.35,0.6,0.9,0.8,0.65,0.8]
oranges = [0.4,0.8,0.9,0.7,0.6,0.8]
```

```python
plt.bar(years,oranges)
```

```python
plt.title('Yield of Oranges')
plt.bar(years,oranges)
plt.plot(years,oranges,'o--r')
```

```python
# bars can also be stacked on top of one another
# use the bottom argument to achieve this
plt.bar(years,apples)
plt.bar(years,oranges,bottom=apples)

plt.legend(['Apple','Orange'])
```

```python
years = range(2000,2006)
apples = [0.4,0.8,0.9,0.7,0.6,0.8]
oranges = [0.4,0.8,0.9,0.7,0.6,0.8]
plt.bar(years,apples)
plt.bar(years,oranges,bottom=apples)
plt.legend(['Apple','Orange'])
```

```python
#to add values to bar chart
years = range(2000,2006)
list1 = []
for i in years:
    list1.append(str(i))
apples = [0.35,0.6,0.9,0.8,0.65,0.8]
oranges = [0.4,0.8,0.9,0.7,0.6,0.8]
def addlabels_apples(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
plt.bar(list1,apples)
addlabels_apples(list1,apples)
```

Bar Plot with Averages

```python
tips_df = sns.load_dataset('tips')
tips_df.head()
```

```python
#We might want to draw a bar chart to visualize how the average bill amount varies across different days of a week
```

```python
#1st way:
bill_avg_df = tips_df.groupby('day')[['total_bill']].mean()
bill_avg_df
```

```python
plt.bar(bill_avg_df.index,bill_avg_df.total_bill)
```

```python
#matplotlib bar plot not fit to get average bar plot
plt.bar(tips_df.day,tips_df.total_bill)
```

```python
#2nd way usimg seaborn plot : calculates the average value automatically
# black lines tells the variation of the values. it is called confidential interval. tells that 50% of the values lie in that line range
sns.barplot(x='day',y='total_bill',data=tips_df)
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df)
```

# Heatmap

```python
# A heatmap is used to visualize 2-dimensional data like a matrix or a table using colors
flights_df = sns.load_dataset("flights")
flights_df.head()
```

```python
plt.plot(flights_df.year,flights_df.passengers)
```

```python
#Using Heatmap
#first create the matrix
flight_mat = flights_df.pivot('month','year','passengers')
```

```python
flight_mat
```

```python
plt.title("Distribution of passengers")
sns.heatmap(flight_mat)
```

```python
sns.heatmap(flight_mat,cmap='Blues',annot=True)
```

```python
sns.heatmap(flight_mat,cmap='Blues',annot=True,fmt='d')
```

# Plotting multiple charts in a grid

Matplotlib and Seaborn also support plotting multiple charts in a grid, using plt.subplots

```python
fig, axes = plt.subplots(2,3, figsize=(16,8)) # plt.subplots return 2 results fig and axes

#Use the axes for plotting
axes[0][0].plot(years,apples,'s-b')
axes[0][0].plot(years,oranges,'o--r')
axes[0][0].set_xlabel('Year')
axes[0][0].set_ylabel('Yield of fruits (in tons)')
axes[0,0].legend(['Apples','Oranges'])
axes[0,0].set_title("Crop yield in Kanto")

#using seaborn scatter plot
axes[0,1].set_title("Sepal length vs width")
sns.scatterplot(x=flower_df.sepal_length,y=flower_df.sepal_width,hue=flower_df.species,s=100,ax=axes[0,1])

#histogram
axes[0,2].set_title('Distribution of sepal width')
axes[0,2].hist([setosa_df.sepal_width,versicolor_df.sepal_width,virginica_df.sepal_width],bins=np.arange(2,5,0.25),stacked=True)
axes[0,2].legend(['Setosa','Versicolor','Virginica'])

#barplor
axes[1,0].set_title('Restaurant bills')
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,ax=axes[1,0])

#heatmap
axes[1,1].set_title('FLight traffic')
sns.heatmap(flight_mat,cmap='Blues',ax=axes[1,1])

#Image
axes[1,2].set_title("Data science meme")
axes[1,2].imshow(img)

plt.tight_layout(pad=2)
```
