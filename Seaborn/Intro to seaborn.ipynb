# Lineplot

```python
sns.lineplot(x='timepoint',y='signal',data=fmri)
plt.show()
```

```python
#adding hue attribute
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event')
plt.show()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event',style='event',markers=True)
plt.show()
```

# Bar Plot

```python
import pandas as pd
pokemon = pd.read_csv("Pokemon.csv")
pokemon.head()
```

```python
sns.set(style='whitegrid')
sns.barplot(x='Legendary',y='Speed',data=pokemon)
plt.show()
```

```python
sns.barplot(x='is_legendary',y='weight_kg',data=pokemon1,hue='generation')
plt.show()
```

# Scatter Plot

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=iris)
plt.show()
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=iris,hue='species',style='species')
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x='sepal_length',y='petal_length',data=iris,hue='petal_length',style='species')
plt.show()
```

# Hist/dist plot

```python
diamonds = sns.load_dataset("diamonds")
diamonds
```

```python
diamonds.shape
```

```python
sns.displot(diamonds['price'])
plt.show()
```

```python
sns.histplot(diamonds['price'],color='green')
plt.show()
```

```python
sns.histplot(diamonds['price'],color='green',bins=10)
plt.show()
```

```python
sns.histplot(diamonds['price'],color='green',bins=10,kde=False)#kde=False:to remove frequency curve. but distplot is deprecated
plt.show()
```

```python
sns.histplot(y=diamonds['price'],color='green')
plt.show()
```

# Joint plot:combination of scatter and histogram

```python
sns.jointplot(x='sepal_length',y='petal_length',data=iris,color='olive')
plt.show()
```

```python
#adding regression line
sns.jointplot(x='sepal_length',y='petal_length',data=iris,color='olive',kind='reg')
plt.show()
```

# Box plot

```python
churn = pd.read_csv('churn.csv')
churn.head()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Churn',y='tenure',data=churn)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='InternetService',y='MonthlyCharges',data=churn)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Contract',y='tenure',data=churn,palette='Set1',linewidth=3)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Contract',y='tenure',data=churn,palette='Set1',order = ['Two year','Month-to-month','One year'])
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Contract',y='tenure',data=churn,palette='Set1',hue='PaymentMethod')
plt.show()
```
